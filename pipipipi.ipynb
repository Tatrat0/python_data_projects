```python
import pandas as pd
import numpy as np
```

```python
df_products = pd.read_csv('Brazilian E-Commerce Public Dataset by Olist\olist_products_dataset.csv')
```

```python
df_products.columns
```

```python
df_order_items = pd.read_csv('Brazilian E-Commerce Public Dataset by Olist\olist_order_items_dataset.csv')
```

```python
df_order_items.columns
```

```python
df_order_reviews = pd.read_csv('Brazilian E-Commerce Public Dataset by Olist\olist_order_reviews_dataset.csv')
```

```python
df_order_reviews.columns
```

```python
df_order_products_reviews_merge = df_order_items[['order_id', 'order_item_id', 'product_id', 'seller_id', 'price']].merge(
           df_products[['product_id', 'product_category_name', 'product_photos_qty']], on='product_id').merge(
               df_order_reviews[['review_id', 'order_id', 'review_score']], on='order_id')
```

```python
df_order_products_reviews_merge.to_csv('Brazilian E-Commerce Public Dataset by Olist/df_order_products_reviews_merge.csv', index=False, sep=";")
```

```python
df_payments = pd.read_csv('Brazilian E-Commerce Public Dataset by Olist/olist_order_payments_dataset.csv')
```

```python
df_payments.columns
```

```python
df_payments_products_merge = df_order_items[['order_id', 'product_id', 'price']].merge(
           df_products[['product_id', 'product_category_name']], on='product_id').merge(
               df_payments[['order_id', 'payment_type',
        'payment_value']], on='order_id')
```

```python
df_payments_products_merge.to_csv('Brazilian E-Commerce Public Dataset by Olist/df_payments_products_merge.csv', index=False, sep=";")
```
